```python
import spacy
NER = spacy.load("en_core_web_sm")
```

```python

from bs4 import BeautifulSoup
import glob
import re
 
# Reading the data inside the xml
# file to a variable under the name
# data
files = glob.glob("./News/*.xml")
docs = []

for file in files:
    with open(file, 'r') as f:
        data = f.read()
    
    # Passing the stored data inside
    # the beautifulsoup parser, storing
    # the returned object
    Bs_data = BeautifulSoup(data, "xml")
    
    # Finding all instances of tag
    # `unique`
    b_unique = Bs_data.find_all('content')
    date = Bs_data.find_all('date')
    
    docs.append({
        'fileName': file.split("/")[-1][:-4],
        'content': b_unique[0].text,
        'date': date[0].text
    })
```

```python
ents = {}

for doc in docs:
    text1= NER(doc['content'])
    for ent in text1.ents:
        if ent.label_ == "PERSON" or ent.label_ == "ORG":
            # ents.append({
            #     'text': ent.text,
            #     'label': ent.label_
            # })
            ents[(ent.text, ent.label_)] = ents.get((ent.text, ent.label_), 0) + 1
```

```python
# files = glob.glob("./News/*.xml")
randomDocs = []
```

```python
def wordCount(text):
    return len(text.split(" "))
```

```python
import random

count = 0
while True:
    if count == 100000:
        print("cannot find results!")
        break
    count += 1
    randomDocs = random.choices(docs, k=9*2)

    firstHalf = 0
    secondHalf = 0
    works = True
    for i, doc in enumerate(randomDocs):
        numWords = wordCount(doc['content'])
        # print(i, doc['fileName'], numWords)
        if numWords > 250:
            works = False

        if i < 9:
            firstHalf += numWords
        else:
            secondHalf += numWords

    if not works:
        continue

    # countinue if the difference is too big
    if abs(firstHalf - secondHalf) > 100:
        continue

    # we need second half to be bigger
    if firstHalf > secondHalf:
        continue

    if set([doc["fileName"] for doc in randomDocs]).isdisjoint(keyDocs):
        break


print(firstHalf, secondHalf)
```
